```python
# required libraries
import pandas as pd
import numpy as np

# visualization libraries
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import PIL
from PIL import Image

# Misc Libaries
import os
import random
import warnings
warnings.filterwarnings('ignore')

# ML/DL libraries
import tensorflow as tf
```

```python
# misc checks
print("Tensorflow Version:", tf.__version__)
print("Num GPUs Available: ", tf.config.list_physical_devices('GPU'))

if tf.test.gpu_device_name():
    print('Default GPU Device: {}'.format(tf.test.gpu_device_name()))
else:
    print("Please install GPU version of TF")
```

```python
# read 25 random each from training and validation images
train_img = []
val_img = []

train_path = '../data/train/images'
val_path = '../data/valid/images'

names = random.choices(os.listdir(train_path), k=25)
for filename in names: 
    full_path = os.path.join(train_path, filename)
    if os.path.isfile(full_path):
        img = Image.open(full_path)
        train_img.append(img)

names = random.choices(os.listdir(val_path), k = 25)
for filename in names:
    full_path = os.path.join(val_path, filename)
    if os.path.isfile(full_path):
        img = Image.open(full_path)
        val_img.append(img)
```

```python
# plot train images
NROWS = 5
NCOLS = 5
plt.figure(figsize = (15,15))
for index, img in enumerate(train_img):
    plt.subplot(NROWS, NCOLS, index + 1)
    plt.imshow(img)
plt.show()
```

```python
# Plot validation images
# plot train images
NROWS = 5
NCOLS = 5
plt.figure(figsize = (15,15))
for index, img in enumerate(val_img):
    plt.subplot(NROWS, NCOLS, index + 1)
    plt.imshow(img)
plt.show()
```
